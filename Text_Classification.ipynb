# **Text Preprocessing**

```python
import re
import string
from nltk.tokenize import word_tokenize

# Fungsi utama untuk membersihkan teks
def clean_text(text):
    # Hapus karakter khusus, mention, hashtag, dan link
    text = text.lower() # convert text to lowercase
    text = re.sub(r"\\t|\\n|\\u|\\", " ", text)  # Hapus tab, newline, dan backslash
    text = text.encode('ascii', 'replace').decode('ascii')  # Hapus karakter non-ASCII
    text = re.sub(r"([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)", " ", text)  # Hapus mention, hashtag, dan link
    text = text.replace("http://", " ").replace("https://", " ")  # Hapus URL yang tidak lengkap
    text = re.sub(r"\d+", "", text)  # Hapus angka
    text = text.translate(str.maketrans("", "", string.punctuation))  # Hapus tanda baca
    text = re.sub(r"\b[a-zA-Z]\b", "", text)  # Hapus karakter tunggal
    text = re.sub(r"\s+", " ", text).strip()  # Hapus spasi ganda dan spasi di awal/akhir

    # Return the cleaned text, or an empty string if it's empty after cleaning
    return text   # This line ensures an empty string is returned instead of None

# Terapkan pembersihan dan tokenisasi pada kolom 'subject'
df['cleaned_text'] = df['subject'].apply(lambda x: word_tokenize(clean_text(x)))
```

# **Remove Stopword**

```python
# Fungsi untuk menghapus stopwords
def rem_stopw(text):
    listStopword = set(stopwords.words('indonesian'))
    return [word for word in text if word not in listStopword]
```

```python
# Terapkan fungsi rem_stopw pada kolom 'cleaned_text'
df['cleaned_text'] = df['cleaned_text'].apply(rem_stopw)
```

# **Word Embedding**

```python
from gensim.models import Word2Vec
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.text import Tokenizer

# Melatih model Word2Vec
word2vec_model = Word2Vec(sentences=df['cleaned_text'].tolist(), vector_size=100, window=5, min_count=1)

# Menyiapkan tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['cleaned_text'].apply(lambda x: ' '.join(x)))  # Menggabungkan kembali list ke string

# Menyiapkan embedding matrix
vocab_size = len(tokenizer.word_index) + 1
embedding_dim = word2vec_model.vector_size
embedding_matrix = np.zeros((vocab_size, embedding_dim))

for word, i in tokenizer.word_index.items():
    if word in word2vec_model.wv:
        embedding_matrix[i] = word2vec_model.wv[word]

# Membuat embedding layer
embedding_layer = Embedding(input_dim=vocab_size,
                            output_dim=embedding_dim,
                            weights=[embedding_matrix],
                            trainable=False)
```
